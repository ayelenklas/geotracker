```python
import streamlit as st
from streamlit_folium import folium_static
import folium
from folium.plugins import FastMarkerCluster
import json
import pandas as pd
```

```python
m = folium.Map(location=[52.520008, 13.404954],
                zoom_start=11,
                prefer_canvas=True,)

with open("../geotracker/website/data/geojson.json") as f:
    file = json.load(f)

samples = pd.read_csv("../geotracker/website/data/r4map.csv")[["lat", "lon", "Address", "Name"]]
```

```python
m.add_child(FastMarkerCluster(samples[["lat", "lon"]], callback=None,
                                                        options=None,
                                                        name=None,
                                                        overlay=True,
                                                        control=True,
                                                        show=None,
                                                        icon_create_function=None))
```
